```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import sys
import PIL
import os
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import umap.umap_ as umap

%matplotlib inline
```

```python
# extracting image data
images = np.load('data.npy')
print(images.shape)
```

```python
# extracting label data
with open('labels.txt') as f:
    labels = f.readlines()
for i in range(len(labels)):
    labels[i] = labels[i][:-1]
print(len(labels))
```

```python
# Multidimensional scaling
mds = MDS(n_components=3)
images_MDS = mds.fit_transform(images)
images_MDS.shape
```

```python
print(images_MDS)
print(type(images_MDS))
np.save('MDS_points.npy', images_MDS)
```

```python
# Principal Component Analysis (currently kernel dies on 16GB of RAM)
pca = PCA(n_components=3)
```

```python
principalComponents = pca.fit_transform(images)
print(principalComponents)
```

```python
# Uniform Manifold Approximation & Projection
fit = umap.UMAP(n_components=3)
%time u = fit.fit_transform(images)
```
